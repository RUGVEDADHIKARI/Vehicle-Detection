```python
while True:
    ret, frame = cap.read()
    if not ret:
        break
    count += 1
    frame = cv2.resize(frame, (1080, 500))
    imgRegion = cv2.bitwise_and(frame, mask)
    results = model.predict(imgRegion)
    a = results[0].boxes.data
    a = a.detach().cpu().numpy()
    px = pd.DataFrame(a).astype("float")
    
    list = []
    detected_classes = [] 
    for index, row in px.iterrows():
        x1 = int(row[0])
        y1 = int(row[1])
        x2 = int(row[2])
        y2 = int(row[3])
        d = int(row[5])
        c = class_list[d]
        if 'car' in c or 'motorcycle' in c or 'truck' in c:
            list.append([x1, y1, x2, y2])
            detected_classes.append(c)
    
    bbox_id = tracker.update(list)
    for i, bbox in enumerate(bbox_id):
        x3, y3, x4, y4, id = bbox
        cx = (x3 + x4) // 2
        cy = (y3 + y4) // 2

        red_line_y = 198
        blue_line_y = 268
        offset = 7
        current_class = detected_classes[i]

        # Update the counters only when the vehicle crosses the lines
        if id not in down and red_line_y < (cy + offset) and red_line_y > (cy - offset):
            down[id] = cy
        if id in down:
            if id not in counter_down and blue_line_y < (cy + offset) and blue_line_y > (cy - offset):
                cv2.circle(imgRegion, (cx, cy), 4, (0, 0, 255), -1)
                cv2.putText(imgRegion, str(id), (cx, cy), cv2.FONT_HERSHEY_COMPLEX, 0.8, (0, 255, 255), 2)
                counter_down.append(id)
                if current_class == 'car':
                    car_counter.append(id)
                elif current_class == 'motorcycle':
                    bike_counter.append(id)
                elif current_class == 'truck':
                    truck_counter.append(id)

        if id not in up and blue_line_y < (cy + offset) and blue_line_y > (cy - offset):
            up[id] = cy
        if id in up:
            if id not in counter_up and red_line_y < (cy + offset) and red_line_y > (cy - offset):
                cv2.circle(imgRegion, (cx, cy), 4, (0, 0, 255), -1)
                cv2.putText(imgRegion, str(id), (cx, cy), cv2.FONT_HERSHEY_COMPLEX, 0.8, (0, 255, 255), 2)
                counter_up.append(id)
                if current_class == 'car':
                    car_counter.append(id)
                elif current_class == 'motorcycle':
                    bike_counter.append(id)
                elif current_class == 'truck':
                    truck_counter.append(id)
        
        text_color = (255, 255, 255)  # white color for text
        red_color = (0, 0, 255)  # (B, G, R)   
        blue_color = (255, 0, 0)  # (B, G, R)
        green_color = (0, 255, 0)  # (B, G, R)

        # Draw lines and text
        cv2.line(imgRegion, (172, 198), (774, 198), red_color, 3)  # Starting coordinates and end of line coordinates
        cv2.putText(imgRegion, 'red line', (172, 198), cv2.FONT_HERSHEY_SIMPLEX, 0.5, text_color, 1, cv2.LINE_AA)
        
        cv2.line(imgRegion, (8, 268), (927, 268), blue_color, 3)  # Second line
        cv2.putText(imgRegion, 'blue line', (8, 268), cv2.FONT_HERSHEY_SIMPLEX, 0.5, text_color, 1, cv2.LINE_AA)    

        cv2.line(frame, (172, 198), (774, 198), red_color, 3)  # Starting coordinates and end of line coordinates
        cv2.putText(frame, 'red line', (172, 198), cv2.FONT_HERSHEY_SIMPLEX, 0.5, text_color, 1, cv2.LINE_AA)
        
        cv2.line(frame, (8, 268), (927, 268), blue_color, 3)  # Second line
        cv2.putText(frame, 'blue line', (8, 268), cv2.FONT_HERSHEY_SIMPLEX, 0.5, text_color, 1, cv2.LINE_AA)  
        
        downwards = len(counter_down)
        cv2.putText(frame, 'going down - ' + str(downwards), (60, 40), cv2.FONT_HERSHEY_SIMPLEX, 0.5, green_color, 1, cv2.LINE_AA)    
    
        upwards = len(counter_up)
        cv2.putText(frame, 'going up - ' + str(upwards), (60, 60), cv2.FONT_HERSHEY_SIMPLEX, 0.5, text_color, 1, cv2.LINE_AA)

    car_counter_navigator = len(set(car_counter))
    cv2.putText(frame, 'Total Cars - ' + str(car_counter_navigator), (60, 80), cv2.FONT_HERSHEY_SIMPLEX, 0.5, green_color, 1, cv2.LINE_AA)

    bike_counter_navigator = len(set(bike_counter))
    cv2.putText(frame, 'Total Motorcycles - ' + str(bike_counter_navigator), (60, 100), cv2.FONT_HERSHEY_SIMPLEX, 0.5, green_color, 1, cv2.LINE_AA)

    truck_counter_navigator = len(set(truck_counter))
    cv2.putText(frame, 'Total Trucks - ' + str(truck_counter_navigator), (60, 120), cv2.FONT_HERSHEY_SIMPLEX, 0.5, green_color, 1, cv2.LINE_AA)
        
    cv2.imshow("Image", frame)
    if cv2.waitKey(1) & 0xFF == 27:  # Exit on ESC key
        break

cap.release()
cv2.destroyAllWindows()
```
